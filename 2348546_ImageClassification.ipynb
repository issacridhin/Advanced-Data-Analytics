```python
# Read the images
img3_masked = cv2.imread('/content/circleblackandwhite.png')
img4_friends = cv2.imread('/content/friends.png')

# Resize the second image to match the first image
img4_resized = cv2.resize(img4_friends, (img3_masked.shape[1], img3_masked.shape[0]))

# Bitwise AND
bitwise_and = cv2.bitwise_and(img3_masked, img4_resized)

# Bitwise OR
bitwise_or = cv2.bitwise_or(img3_masked, img4_resized)

# Bitwise XOR
bitwise_xor = cv2.bitwise_xor(img3_masked, img4_resized)

# Bitwise NOT
bitwise_not = cv2.bitwise_not(img3_masked)
```

**Bitwise AND**

```python
cv2_imshow(bitwise_and)
```

**Bitwise OR**

```python
cv2_imshow(bitwise_or)
```

**Bitwise XOR**

```python
cv2_imshow(bitwise_xor)
```

**Bitwise NOT**

```python
cv2_imshow(bitwise_not)
```

**Image PreProcessing**

```python
# Load the image
img = cv2.imread('/content/tom.png')

# Resizing
resized_img = cv2.resize(img, (400, 400))

# Normalization
normalized_img = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

# Data Augmentation (flipping)
flipped_img = cv2.flip(img, 1)

# Cropping
cropped_img = img[0:100, 0:100]

# Grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Histogram Equalization
equ_img = cv2.equalizeHist(gray_img)
```

**Resizing Image**

```python
cv2_imshow(resized_img)
```

**Normalization MinMax**

```python
cv2_imshow(normalized_img)
```

**Flipped Image**

```python
cv2_imshow(flipped_img)
```

**Cropped Image**

```python
cv2_imshow(cropped_img)
```

**Gray Image**

```python
cv2_imshow(gray_img)
```

**Histogram Equalization**

```python
cv2_imshow(equ_img)
```

```python
# Read the image
img_jerry = cv2.imread('/content/jerry.png')
img_jerry_resized = cv2.resize(img_jerry, (300,200))

# Gaussian Blurring
gaussian_blur_img = cv2.GaussianBlur(img_jerry_resized, (5, 5), 0)

# Image Rotation
rotated_img = cv2.rotate(img_jerry_resized, cv2.ROTATE_90_CLOCKWISE)

# Noise Reduction
denoised_img = cv2.fastNlMeansDenoisingColored(img_jerry_resized, None, 10, 10, 7, 21)

# Edge Detection
edges_img = cv2.Canny(img_jerry_resized, 100, 200)

# Color Space Conversion
hsv_img = cv2.cvtColor(img_jerry_resized, cv2.COLOR_BGR2HSV)

# Contrast Adjustment
adjusted_img = cv2.convertScaleAbs(img_jerry_resized, alpha=1.5, beta=0)
```

**Gaussian Blur**

```python
cv2_imshow(gaussian_blur_img)
```

**Image Rotation**

```python
cv2_imshow(rotated_img)
```

**Noise Reduction**

```python
cv2_imshow(denoised_img)
```

**Edge Detection**

```python
cv2_imshow(edges_img)
```

**Color Space Conversion**

```python
cv2_imshow(hsv_img)
```

**Contrast Adjustment**

```python
cv2_imshow(adjusted_img)
```
